```python
import numpy as np # defines np as referring to NumPy and imports NumPy for use
import matplotlib.pylab as plt # defines plt as representing matplotlib.pylab and makes usable for following code
import librosa # convert or describe audio file in terms usable and representable by python code
import librosa.display # enables display of items imported using librosa
```

```python
def plotAudio(filename,sampleRate): #defining function "plotAudio" as having two arguments 1) filename and 2) sampleRate#
    sig , sr = librosa.load(filename,mono=True,sr=sampleRate) #librosa.load loads the file used for argument filename. mono=True yeilds mono input for audio file, sr=sampleRate defaults to file sample rate?#
    time=np.arange(0,sig.shape[0])/sr #np is NumPy and is used for arrays.  
    xlabel = 'Time (seconds)' 
    ylabel = 'Amplitude'
    excursion=max(abs(sig)) # ?
    sigNew = sig / excursion # ? 
    plt.plot(time,sigNew) #plt is matplotlib and plot is a function, here with arguments time(x) and sigNew(y), where sigNew is the file imported using librosa 
    plt.xlabel(xlabel) #tells matplotlib to label the x and y axes according to prior definition in function
    plt.ylabel(ylabel)
```

```python
plotAudio('avm.wav',44100)
```

```python
plotAudio('avm.wav',441)
```
